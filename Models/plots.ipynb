```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mode=4
perturbation="1.0000" 

central_IceModel="/home/anegi/multisim/ppc/ice/spice_ftp-v2/icemodel_.dat"
Perturbed_IceModels="/data/user/anegi/ppc/models/"

data=np.loadtxt(central_IceModel) #
data_Amp=np.loadtxt(str(Perturbed_IceModels)+"/Amp/"+str(mode)+"/Amp_"+str(mode)+"_"+str(perturbation)+".dat")
data_Phs=np.loadtxt(str(Perturbed_IceModels)+"/Phs/"+str(mode)+"/Phs_"+str(mode)+"_"+str(perturbation)+".dat")

depth=data[:,0]
abs_A=data_Amp[:,1]
abs_P=data_Phs[:,1]
abs_=data[:,1]

plt.plot(depth,abs_A,color='red',label='Amp')
plt.plot(depth,abs_P,color='black',label='Phase')
plt.plot(depth,abs_,color='grey',linestyle='dashed',label='central')
plt.legend()
```
